Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(df[['Sex']])
df['Sex']=le.transform(df['Sex'])
```

```python
le=LabelEncoder()
le.fit(df["Embarked"])
df["Embarked"]=le.transform(df["Embarked"])
```

Spilliting the data into test and train

```python
x=df.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
y=df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=5)
```

Applying Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_test = lr.predict(x_test)
y_pred_train = lr.predict(x_train)
acc_test = accuracy_score(y_test, y_pred_test)
acc_train = accuracy_score(y_train,y_pred_train)
print(f"testing score is :{acc_test}")
print(f"training score is :{acc_train}")
```

Conclusion

```python
#The close alignment between training and testing scores suggests consistency in model performance, reducing concerns about overfitting.
#The high testing score indicates that the model is reliable and effective for making predictions on new data.
#
```
